# LVK data downloader

A tool for downloading LVK noise-only and 'blip' glitch data.

Note that after the event and glitch catalogs are downloaded (the first time this is run), the downloads take about ~1 min per dataset.

So for ~100 events this will take more than an hour, unless this is run in parallel.
